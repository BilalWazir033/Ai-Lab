```python
from collections import deque
```

```python
graph = {
    'A': ['B','C'],
    'B': ['D','E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
```

```python
def bfs_search(graph,start, target):
    visited=set() # if a node is visited come here
    queue=deque() # 

    visited.add(start) #
    queue.append(start)# 

    while queue: # how many elements in queue it will run
        node=queue.popleft() # the first node come here in this case is 'A'

        if node== target: # match
            print("Target found: ",node) # 
            return True # means terminat here
        for neighbor in graph[node]:# neighber of visited element here at node will come 'A' at neighber is B, C
            if neighbor not in visited: # 
                visited.add(neighbor) # 
    print("Target not found") # 
    return False # 
```

```python
bfs_search(graph,'A', 'F')
```
