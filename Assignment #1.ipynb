### 1.3 Indexing, Slicing, Boolean Indexing

```python
# Step 1: Create 6×6 matrix of random integers (0–99)
import numpy as np

A = np.random.randint(0, 100, (6, 6))
print("Original Matrix:\n", A)
```

```python
#Step 2: Replace every value > 50 with 999
A[A > 50] = 999
print("\nAfter replacing > 50 with 999:\n", A)
```

```python
# Step 3: Extract Main Diagonal
diag = np.diag(A)
print("\nMain Diagonal:\n", diag)
```

```python
# Step 4: Elements Strictly Above the Diagonal
np.triu(A, k=1)
```

```python
above_diag = np.triu(A, k=1)
print("\nElements Above Diagonal:\n", above_diag)
```

```python
above_values = A[np.triu_indices(6, k=1)]
print("\nOnly Above-Diagonal Values:\n", above_values)
```

```python
# Step 5: Row Sums and Column Sums
row_sums = np.sum(A, axis=1)
col_sums = np.sum(A, axis=0)

print("\nRow Sums:\n", row_sums)
print("\nColumn Sums:\n", col_sums)
```

```python
# Bonus: Replace Every Even Number with Its Negative
A % 2 == 0
```

```python
A[A % 2 == 0] *= -1
print("\nAfter replacing even numbers with negative:\n", A)
```

## Part 2: Pandas – Data Wrangling & Analysis

```python
# First: Load the Dataset
import pandas as pd

url = "https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv"
df = pd.read_csv(url)

print(df.head())

```

```python
# Show 5 Cars with Highest and 5 with Lowest mpg
highest_mpg = df.sort_values('mpg', ascending=False).head(5)
print("Highest MPG:\n", highest_mpg)
```

```python
lowest_mpg = df.sort_values('mpg', ascending=True).head(5)
print("Lowest MPG:\n", lowest_mpg)
```

```python
# Filter Cars (6 or 8 cylinders AND hp > 150)
filtered = df[(df['cyl'].isin([6,8])) & (df['hp'] > 150)]
print(filtered)

```

```python
# Create New Column: power_to_weight = hp / wt
df['power_to_weight'] = df['hp'] / df['wt']
print(df[['hp','wt','power_to_weight']].head())
```

```python
# Mean mpg Grouped by Number of Cylinders
mean_mpg = df.groupby('cyl')['mpg'].mean()
print(mean_mpg)

```

### 2.4 Grouping, Aggregation, Pivot Tables

```python
import pandas as pd

url = "https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv"
df = pd.read_csv(url)
```

```python
# Average
grouped = df.groupby(['cyl', 'gear'])[['mpg', 'hp', 'wt']].mean()
print("Average mpg, hp, wt by cyl and gear:\n", grouped)

```

```python
pivot = pd.pivot_table(df, values='mpg', index='cyl', columns='gear', aggfunc='mean')
print("\nPivot Table (Mean MPG by cyl vs gear):\n", pivot)
```

```python
# Bonus: Car with best hp / mpg ratio
df['hp_per_mpg'] = df['hp'] / df['mpg']
most_efficient = df.loc[df['hp_per_mpg'].idxmin()]
print("\nCar with best hp/mpg ratio:\n", most_efficient)
```

## Part 3: Matplotlib – Visualization Fundamentals

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load mtcars dataset
url = "https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv"
df = pd.read_csv(url)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14,10))
```

```python
scatter = axs[0,0].scatter(df['hp'], df['mpg'], c=df['wt'], cmap='viridis', s=70, alpha=0.8)
axs[0,0].set_xlabel('HP')
axs[0,0].set_ylabel('MPG')
axs[0,0].set_title('MPG vs HP (colored by Weight)')
plt.colorbar(scatter, ax=axs[0,0], label='Weight')
```

```python
avg_mpg = df.groupby('cyl')['mpg'].mean()
axs[0,1].bar(avg_mpg.index, avg_mpg.values, color='coral')
axs[0,1].set_xlabel('Cylinders')
axs[0,1].set_ylabel('Average MPG')
axs[0,1].set_title('Average MPG by Cylinder Count')
```

```python
axs[1,0].boxplot([df[df['am']==0]['mpg'], df[df['am']==1]['mpg']], labels=['Auto','Manual'])
axs[1,0].set_title('MPG by Transmission')
axs[1,0].set_ylabel('MPG')
```

```python
x = df['qsec']
y = df['mpg']

axs[1,1].scatter(x, y, color='skyblue', s=70, alpha=0.7)
# Linear trend line
m, b = np.polyfit(x, y, 1)  # degree 1
axs[1,1].plot(x, m*x + b, color='red', linewidth=2)
axs[1,1].set_xlabel('Quarter Mile Time (qsec)')
axs[1,1].set_ylabel('MPG')
axs[1,1].set_title('MPG vs QSec with Trend Line')
```

```python
plt.tight_layout()
plt.show()
```

## Part 4: Mini-Project – mtcars EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load mtcars dataset
url = "https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv"
df = pd.read_csv(url)
```

```python
# Step 1: Create new columns
# 1) Convert mpg to km/liter
df['km_per_liter'] = df['mpg'] * 0.425144

# 2) Performance index: hp / wt
df['performance_index'] = df['hp'] / df['wt']

print(df[['mpg','km_per_liter','hp','wt','performance_index']].head())
```

```python
# Step 2a: Which cylinder count has highest average performance_index?
avg_perf = df.groupby('cyl')['performance_index'].mean()
print("Average performance index by cylinder:\n", avg_perf)

# Plot
plt.figure(figsize=(6,4))
avg_perf.plot(kind='bar', color='skyblue')
plt.xlabel('Cylinders')
plt.ylabel('Average Performance Index')
plt.title('Performance Index by Cylinder Count')
plt.show()
```

```python
# Step 2b: Relationship between weight (wt) and acceleration (qsec)
plt.figure(figsize=(6,4))
plt.scatter(df['wt'], df['qsec'], color='coral', s=70, alpha=0.7)
m, b = np.polyfit(df['wt'], df['qsec'], 1)
plt.plot(df['wt'], m*df['wt'] + b, color='blue', linewidth=2)
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('Quarter Mile Time (qsec)')
plt.title('Weight vs Acceleration (qsec)')
plt.show()
```

```python
# Step 2c: Compare Automatic vs Manual Transmission on mpg and hp
fig, axs = plt.subplots(1, 2, figsize=(12,5))

# Boxplot: MPG by transmission
axs[0].boxplot([df[df['am']==0]['mpg'], df[df['am']==1]['mpg']], labels=['Auto','Manual'])
axs[0].set_title('MPG by Transmission')
axs[0].set_ylabel('MPG')

# Boxplot: HP by transmission
axs[1].boxplot([df[df['am']==0]['hp'], df[df['am']==1]['hp']], labels=['Auto','Manual'])
axs[1].set_title('HP by Transmission')
axs[1].set_ylabel('Horsepower')

plt.tight_layout()
plt.show()
```

```python
# Step 2d (Bonus): Correlation heatmap of numeric columns
# Select only numeric columns
numeric_df = df.select_dtypes(include=np.number)

# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Columns')
plt.show()

```
