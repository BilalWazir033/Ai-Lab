## Fetching rows and columns using iloc 

```python

data.iloc[0] # Important for rows as well columns 
# column name and value for each column of first row
data.iloc[23]
data.iloc[1:4] # 1,2,3 will be printed
data.iloc[1:6:2] # from on to 6 and even rows only
data.iloc[[1,2,5]] # specific by fancing indexing
```

```python
# for columns now
data.iloc[1:4,2:5] # means 1 to 3rd row and 2 to 4th column
data.iloc[[1,2,4,5],[4,5,10]] # for row 1,2,4,5 and column 4,5,10 excellent yar
```

## Filtering dataframe on a condition | Masking using boolean

```python
mask=data['city']=='Rajkot' 
data[mask] # only matches played in rajkot
data[mask].shape[0] # but how many matches played
```

```python
# through function
def get_city_match(city):
    mask=data['city']==city 
    return data[mask].shape[0]
get_city_match('Pune')
# now its very easy and can get dynamically
```

```python
# lets ciyt with year
mask=(data['city']=='Hyderabad') & (data['season']==2012)
data[mask].shape[0] # means 4 matches in 2012 and in 
```

```python
# using function
def city_year_match(city,year):
    mask=(data['city']==city) & (data['season']==year) # we used & operator not 'and' b/c here is boolean 
    return data[mask]
bro=city_year_match('Hyderbad',2012)
bro.iloc[:,[1,2,4,5,10]] 
```

```python
# another way
mask1=data['city']=='Hyderabad'
mask2=data['season']==2012

data[mask1 & mask2] # by putting .shape[0] at the end will give the number 
```

## The value_counts() function
- find the frequency column of data in one column
- its work on series i.e categorical data column wise only
- the row is the numerical data which is unique


```python
data['winner'].value_counts()
data['season'].value_counts()
data[['winner','season']].value_counts()
```

## The sort_values() method

```python
# top most 5 teams with most matches
(data['team1'].value_counts()+data['team2'].value_counts()).sort_values(ascending=False).head () # sort the values in ascending order
```

```python
data.sort_values(['city','season'],ascending=[True,False]).tail()
```
