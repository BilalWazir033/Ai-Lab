## Filtering dataframe on a condition | Masking using boolean

```python
mask=data['city']=='Rajkot' 
data[mask] # only matches played in rajkot
data[mask].shape[0] # but how many matches played
```

```python
# through function
def get_city_match(city):
    mask=data['city']==city 
    return data[mask].shape[0]
get_city_match('Pune')
# now its very easy and can get dynamically
```

```python
# lets ciyt with year
mask=(data['city']=='Hyderabad') & (data['season']==2012)
data[mask].shape[0] # means 4 matches in 2012 and in 
```

```python
# using function
def city_year_match(city,year):
    mask=(data['city']==city) & (data['season']==year) # we used & operator not 'and' b/c here is boolean 
    return data[mask]
bro=city_year_match('Hyderbad',2012)
bro.iloc[:,[1,2,4,5,10]] 
```

```python
# another way
mask1=data['city']=='Hyderabad'
mask2=data['season']==2012

data[mask1 & mask2] # by putting .shape[0] at the end will give the number 
```

```python
# now vk highest runs against team 
vk=delivery[delivery['batsman']=='V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
# now make a function for it
def run_by_batsman(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3) 
    # return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0] # it will give the one against made high score
```

```python
batsman=run_by_batsman('CH Gayle')
batsman
```

### Goal
- Find the most destructive death over batsman in the history of IPL
- Strike Rate=(Number of runs/Number of balls)*100
- min batsman 200 balls in over 16-20

```python
death_over=delivery[delivery['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
# all_batsman[x].sort_values(ascending=False)
batsman_list=all_batsman[x].index.to_list()
```

```python
# runs scored by all these 43 batsman
# balls played by these 43 batsman
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
strike_rate=(runs/balls)*100
strike_rate.sort_values(ascending=False).head()
```

## Pivot Table

```python
mask=delivery['batsman_runs']==6
six=delivery[mask]
six.shape
```

```python
sixes=six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
sixes.head()
```
