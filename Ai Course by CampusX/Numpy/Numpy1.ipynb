#  Indexing with Boolean Arrays

```python
arr=np.random.randint(low=1, high=100,size=20).reshape(5,4)
arr
```

```python
arr[4][3] # in row 4 element 3rd
```

```python
arr>50
```

```python
arr[arr>50]
```

```python
arr[(arr>50) & (arr%2!=0)] # arr >50 and is odd
```

```python
arr[(arr>50) & (arr%2!=0)]=0 # initialize that with zero
```

```python
arr
```

# Broadcasting
- In numpy we can perform operation on disimilar arrays through broadcasting




```python
a3=np.arange(9).reshape(3,3)
a4=np.arange(3).reshape(1,3)
print(a3)
a4
```

```python
a3+a4 # addition will done in shuch a way that a4 add to each row of a3 
# further 3,3 > 1,3 means 3,3 match so possible
```

```python
a5=np.arange(3).reshape(1,3)
a6=np.arange(12).reshape(4,3)
# operation will perform 1,3 .. 4,3 if one then check its right with the 
# right below element if match operation can be purformed
a5=np.arange(3).reshape(3,1)
a6=np.arange(12).reshape(3,4)
```

```python
a5=np.arange(3).reshape(1,3) # no operation performed
a6=np.arange(12).reshape(4,4)
```

```python
a5=np.arange(3).reshape(1,3)
a6=np.arange(12).reshape(4,4)
```

```python
a5=np.arange(3).reshape(1,3)
a6=np.arange(12).reshape(3,1) # dignol are 1 then non dignal should same in self
```
