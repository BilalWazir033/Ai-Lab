# Numpy Array Operations

### 1. Basic Operation


```python
arr13=np.array([1,2,3,4,5,6])
arr14=np.array([4,5,6,7,8,9])
print(arr13,"\n",arr14)
```

```python
arr13-arr14
arr13*2  # multiply 2 to each elemnet 
arr13>2
```

```python
arr3=np.arange(6).reshape(2,3) # 6 elements , 2 rows and 3 columns
arr4=np.arange(6,12).reshape(3,2)
print(arr3,"\nlal\n",arr4)
```

```python
arr3.dot(arr4) # here 2,3=3,2 so a 2,2 matrix will be formed
```

### 2. Unary Operation

```python
arr4.max()
arr3.min()
```

```python
arr4.max(axis=0) # row wise maximum
```

```python
arr3.min(axis=1) # column wise min
# do not need object direct call the class
```

### 3. Universal Function

```python
np.median(arr4)
```

```python
np.sin(arr3)
```

#  Indexing with Boolean Arrays

```python
arr=np.random.randint(low=1, high=100,size=20).reshape(5,4)
arr
```

```python
arr[4][3] # in row 4 element 3rd
```

```python
arr>50
```

```python
arr[arr>50]
```

```python
arr[(arr>50) & (arr%2!=0)] # arr >50 and is odd
```

```python
arr[(arr>50) & (arr%2!=0)]=0 # initialize that with zero
```
