```python
# now check the faster 
import time #to measure speed
```

```python
# normal python code without library
x=range(10000000)
y=range(10000000, 20000000)
start_time=time.time() # time function give the current time
c=[(x+y) for x,y in zip(x,y)] # t|ake each element of every list and combine it and then add it to a new list
print(time.time()-start_time) # sub the start time that is before starting the function
```

```python
# now we do for numpy
a=np.arange(10000000)
b=np.arange(10000000, 20000000)
start_time=time.time()
c=a+b
print(time.time()-start_time)
```

 # Indexing, Slicing and Iteration

```python
arr12=np.arange(24).reshape(6,4)
arr12
```

```python
arr12[2:5]
arr12[5:6,1:3]
arr12[:,1:3]
```

```python
arr12[:,2] # column 3rd 
```

```python
arr12[4:6,1:3] # 5th and 6th column, 2nd and 3rd column
```

```python
for i in arr12:
    print(i)
```

```python
for i in np.nditer(arr12):
    print(i)
```
