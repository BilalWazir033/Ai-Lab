# Libraries
## NumPy?
**NumPy** stands for **Numerical Python**.

**Uses:**
- Numerical computing
- Scientific computing
- Data science
- Machine Learning
- Artificial Intelligence

It provides **fast and efficient operations** on large datasets.

## Why NumPy is Important in AI?
**Importance:**
- AI works with **numbers and matrices**
- NumPy is **faster than Python lists**
- Used internally by **Pandas, TensorFlow, PyTorch**

**AI models use:**
- Vectors
- Matrices
- Multi-dimensional arrays

All handled efficiently by NumPy.

## NumPy Array
**Definition:** A NumPy array is a collection of elements of the same data type.

**Types:**
- 1D array (Vector)
- 2D array (Matrix)
- 3D or more (Tensor)

## Creating a NumPy Array
```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```
Output:
```
[1 2 3 4 5]
```

# Array vs Python List
| Feature | Python List | NumPy Array |
|------|-----------|------------|
| Speed | Slow | Fast |
| Data Type | Mixed | Same |
| Memory | More | Less |
| Operations | Limited | Advanced |

## Checking Array Properties
```python
arr = np.array([10, 20, 30])
print(arr.ndim)   # Dimensions
print(arr.shape)  # Shape
print(arr.size)   # Total elements
print(arr.dtype)  # Data type
```
## Creating Arrays Using Built-in Functions
```python
np.zeros(5)
np.ones(5)
np.arange(1, 10)
np.linspace(1, 10, 5)
```
**Explanation:**
- `zeros()` → All zeros
- `ones()` → All ones
- `arange()` → Range of numbers
- `linspace()` → Evenly spaced numbers

## 2D Arrays (Matrices)
```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix)
```
**Rows = 2, Columns = 3**

## Accessing Array Elements
```python
arr = np.array([10, 20, 30, 40])
print(arr[0])
print(arr[-1])
```
2D Access:
```python
matrix[0][1]
```


## Slicing in NumPy
```python
arr = np.array([1, 2, 3, 4, 5])
print(arr[1:4])
```
**Format:** `[start : end]`

## Mathematical Operations
```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```
Element-wise operations.

## Universal Functions (ufuncs)
```python
np.sqrt(arr)
np.sum(arr)
np.mean(arr)
np.max(arr)
np.min(arr)
```

## Reshaping Arrays
```python
arr = np.array([1,2,3,4,5,6])
new_arr = arr.reshape(2,3)
print(new_arr)
```
Reshape changes **structure only**, not data.

## Broadcasting
```python
arr = np.array([1, 2, 3])
print(arr + 5)
```
Result: `[6 7 8]`


## Copy vs View
```python
a = np.array([1, 2, 3])
b = a.copy()
c = a.view()
```
- **Copy → New memory**
- **View → Same memory**

## Random Numbers
```python
np.random.rand(3)
np.random.randint(1, 10, 5)
```
Used in AI for **weights initialization**.

## Sorting and Searching
```python
np.sort(arr)
np.where(arr > 3)

---
# Pandas?
**Pandas** is a Python library used for **data analysis and data manipulation**.

It helps us:
- Read data
- Clean data
- Analyze data
- Prepare data for AI & ML

Pandas works on top of **NumPy**.

## Why Pandas is Important in AI?
**Importance:**
- AI needs **clean and structured data**
- Pandas helps in **data preprocessing**
- Used before training ML models

Used for:
- Handling missing values
- Feature selection
- Data transformation

## Installing Pandas
**Install Pandas:**
```bash
!pip install pandas
```

**Import Pandas:**
```python
import pandas as pd  # pd is alias for pandas
```

## Core Data Structures in Pandas
Pandas has two main data structures:

### 1. Series
- One-dimensional
- Like a column

### 2. DataFrame
- Two-dimensional
- Like a table (rows & columns)

## Pandas Series
**Definition:** A Series is a one-dimensional labeled array.

```python
import pandas as pd
s = pd.Series([10, 20, 30, 40])
print(s)
```

## Pandas DataFrame
**Definition:** A DataFrame is a 2D table of data.

```python
data = {
    'Name': ['Ali', 'Sara', 'Ahmed'],
    'Age': [20, 22, 21]
}

df = pd.DataFrame(data)
print(df)
```

## Reading Data from Files
```python
pd.read_csv('data.csv')
pd.read_excel('data.xlsx')
pd.read_json('data.json')
```

Used to load datasets for AI.

## Viewing Data
```python
df.head()
df.tail()
df.info()
df.describe()
```

- `head()` → First rows
- `tail()` → Last rows
- `info()` → Structure
- `describe()` → Statistics

## Selecting Columns
```python
df['Age']
df[['Name', 'Age']]
```
## Selecting Rows
```python
df.loc[0]
df.iloc[1]
```

- `loc` → Label based
- `iloc` → Index based

## Filtering Data
```python
df[df['Age'] > 20]
```

Filtering data means **selecting only useful rows** from a dataset based on a condition.

- In AI and Machine Learning, raw data is often **large and noisy**.
- We usually need only **relevant records** for training models.

- `df['Age'] > 20` creates a **Boolean condition** (True/False)
- Pandas keeps only rows where the condition is **True**

## Adding New Column
```python
df['Grade'] = ['A', 'B', 'A']
```

Adding a new column is used to **store new information or features**.

- A column can be:
  - Calculated values
  - Labels (target/output)
  - Categories or scores

## Removing Columns
```python
df.drop('Grade', axis=1, inplace=True)
```
Removing columns is used to **delete unnecessary or useless features**.

- `axis=1` → column-wise operation
- `inplace=True` → changes original DataFrame

## Handling Missing Values
```python
df.isnull()
df.dropna()
df.fillna(0)
```
Missing values are very common in real-world data.

### Functions
- `isnull()` → checks missing values
- `dropna()` → removes rows with missing data
- `fillna()` → replaces missing values

## Sorting Data
```python
df.sort_values('Age')
```
Sorting means arranging data in **ascending or descending order**.

## GroupBy Operation
```python
df.groupby('Grade').mean()
```

GroupBy is used to **split data into groups and apply aggregation**.

### Explanation
- Groups data based on a column
- Applies functions like:
  - mean()
  - sum()
  - count()

## Applying Functions
```python
df['Age'].apply(lambda x: x + 1)
```
The `apply()` function applies a **custom function** to data.

### Explanation
- `lambda` is an anonymous function
- Applied to each value in the column

## Merging DataFrames
```python
pd.merge(df1, df2, on='id')
```
Merging combines two DataFrames into one based on a **common key**.

## Part 3: Matplotlib – Visualization Fundamentals

### 3.1 Basic Plots

```python
plt.figure(figsize=(10,6))
plt.hist(df['mpg'], bins=15, color='skyblue', edgecolor='k', alpha=0.7)
plt.xlabel('Miles per Gallon'); plt.ylabel('Count')
plt.title('MPG Distribution – 1974 Cars')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.scatter(df['wt'], df['mpg'],
            c=df['hp'], cmap='viridis', s=70, alpha=0.8)
plt.colorbar(label='Horsepower')
plt.xlabel('Weight (1000 lbs)'); plt.ylabel('MPG')
plt.title('Weight vs Fuel Efficiency')
plt.show()
```

### 3.2 Subplots & More Styles

```python
fig, axs = plt.subplots(2, 2, figsize=(12,9))

axs[0,0].scatter(df['hp'], df['mpg'], alpha=0.6)
axs[0,0].set_title('HP vs MPG')

axs[0,1].hist(df['wt'], bins=12, color='coral')
axs[0,1].set_title('Weight Distribution')

axs[1,0].boxplot([df[df.cyl==4].mpg, df[df.cyl==6].mpg, df[df.cyl==8].mpg],
                 labels=['4','6','8'])
axs[1,0].set_title('MPG by # Cylinders')
axs[1,0].set_ylabel('MPG')

axs[1,1].bar(df.gear.value_counts().index, df.gear.value_counts())
axs[1,1].set_title('Cars per Gear Count')

plt.tight_layout()
plt.show()
```
